# Question 3 : What makes people happy in the year 2019

```python
data5.head()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 6),
                         sharey='row', sharex='row') #we created a grid of 6 plots

axes[0, 0].scatter(x=data5['Social support'],y=data5['Score']) # plot at 1st row, 1st col and 1st position
axes[0, 0].set_xlabel('Social support')
axes[0, 0].set_ylabel('Score')

axes[0, 1].scatter(x=data5['Healthy life expectancy'],y=data5['Score'])# plot at 1st row, 2st col and 2st position
axes[0, 1].set_xlabel('Healthy life expectancy')
axes[0, 1].set_ylabel('Score')

axes[1, 0].scatter(x=data5['Freedom to make life choices'],y=data5['Score'])# plot at 2nd row, 1st col and 1st position
axes[1, 0].set_xlabel('Freedom to make life choices')
axes[1, 0].set_ylabel('Score')

axes[1, 1].scatter(x=data5['Generosity'],y=data5['Score'])# plot at 2nd row, 2nd col and 2nd position
axes[1, 1].set_xlabel('Generosity')
axes[1, 1].set_ylabel('Score')

axes[2, 0].scatter(x=data5['Perceptions of corruption'],y=data5['Score'])# plot at 3rd row, 1st col and 1st position
axes[2, 0].set_xlabel('Perceptions of corruption')
axes[2, 0].set_ylabel('Score')

axes[2, 1].scatter(x=data5['GDP per capita'],y=data5['Score'])# plot at 3rd row, 2nd col and 2nd position
axes[2, 1].set_xlabel('GDP per capita')
axes[2, 1].set_ylabel('Score')




fig.tight_layout()
```

### From the above subplots, we can see that, except Perception of corruption, other things like social support, health, life expectancy, Freedom to to make life choices, Generosity, GDP per Capita makes people happy.
